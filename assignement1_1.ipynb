```python
import warnings
import keras
import numpy as np
import bqplot 
import pandas as pd

def average(x,y,z):
    av = (x + y +z) / 3
    return round(av,3)

def MyFunc(x):
    return str(round(x))

def func3(element):
    list = []
    setList=  set()
    dict = {}
    list.append(element)
    setList.add(element)
    dict[element] = element
    tuple = (list,setList,dict)
    return tuple

def fin_element_tuple(listElement,element):
     return listElement.count(element)

def odd_dict(list):
    dict_of_list = {}
    for i in range(1,len(list),2):
        dict_of_list[str(i)] = list[i]
    return dict_of_list
    
class BankAccount: 
   """Bank Account protected by a account number and pin""" 
   def __init__(self, account_number, pin, initial_balance): 
        """Initial pin is 'pin'.""" 
        self.i = initial_balance 
        self.a = account_number
        self.p = pin
   def deposit(self, pin, amount): 
        """Increment account balance by amount and return new balance.""" 
        if self.p == pin :
           self.i += amount
        return self.i
   def get_balance(self, pin): 
    """Return account balance.""" 
    if pin == self.p:
        return self.i 

def addEnd_array(arr,element):
    return np.append(arr, element)    

def sum_multiple_of(arr):
    arr_4_7 = np.copy(np.select([arr%4==0,arr%7==0],[arr,arr]))
    return np.sum(arr_4_7)

def reverse_numpy_array(arr):
   return arr[::-1]

def most_frequent_value(arr):
    binCountArray = np.bincount(arr)
    most_occured_number= binCountArray.argmax()
    return most_occured_number

def numpy_array_to_pandas_Serie(np_array):
   return pd.Series(np_array)

def fillnan(df):
    return df.fillna(1)

def extract_data(DataFrame,col_start,col_end,row_start,row_end):
   extract_DataFram = DataFrame.iloc[row_start:row_end, col_start:col_end]
   return extract_DataFram

def inner_join(DataFram1,DataFram2,column):
    return pd.merge(DataFram1,DataFram2,on=column)

def count_nan_column(DataFram , column):
      return DataFram[column].isna().sum().sum() 
```
